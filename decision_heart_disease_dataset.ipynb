```python
data=pd.get_dummies(data,columns=['sex','cp','fbs','restecg','exang','slope','ca','thal'])
data
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
col=['age','trestbps','chol','thalach','oldpeak']
data[col]=sc.fit_transform(data[col])
#data[col]=sc.fit(data[col])
data
```

```python
x=data.drop(columns='target')
y=data['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_score=[]

for i in range(1,len(x.columns)+1):
    dt_class=DecisionTreeClassifier(max_features=i,random_state=1)    #default is gini impurity
    dt_class.fit(x_train,y_train)
    dt_score.append(dt_class.score(x_test,y_test))
```

```python
plt.plot([i for i in range(1,len(x.columns)+1)],dt_score,color='red')
```

```python
plt.figure(figsize=(15,5))
plt.plot([i for i in range(1,len(x.columns)+1)],dt_score,color="red")
for i in range(1,len(x.columns)+1):
    plt.text(i,dt_score[i-1],(i,round(dt_score[i-1],2)))
```

```python
plt.figure(figsize=(15,5))
plt.plot([i for i in range(1,len(x.columns)+1)],dt_score,color="red")
for i in range(1,len(x.columns)+1):
    plt.text(i,dt_score[i-1],(i,round(dt_score[i-1],2)))
    
plt.xticks([i for i in range(1, len(x.columns) + 1)])
plt.xlabel("Max Features")
plt.ylabel("Scores")
plt.title("Decision Tree Classifier")
plt.show()    
```

```python
y_pred=dt_class.predict(x_test)
y_pred
```

```python
y_test.values
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)
```
